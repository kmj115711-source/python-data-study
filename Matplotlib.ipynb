## 8. 맷플롯립(Matplotlib)
데이터를 시각화하는 모듈로, 그래프와 차트를 제공

```python
# 필요한 모듈 가져오기
import numpy as np
# 맷플롯립은 여러 모듈로 구성됨
# 그 중 그래프를 제공하는 모듈 꺼내기
import matplotlib.pyplot as plt
```

```python
# 그래프를 그릴 때 한글이 깨지는 현상이 발생 =>
from matplotlib import rc
rc('font', family='Malgun Gothic')
```

### 8.1 그래프 그리기
#### 선 그래프
- plot : 선 그래프를 그리는 함수

```python
# 먼저 그래프에 담을 데이터 생성
a = np.arange(1,11)
a
# 숫자 배열을 그래프에 그리기
plt.plot(a)
# x축 : 우리가 담은 데이터
# y축 : plt가 자동으로 설정함
```

#### 막대 그래프
- bar : 화면에 막대 그래프를 그리는 함수

```python
# 그래프에 담을 데이터 생성
# 학급별 평균 시험 점수
a = ['1반','2반','3반','4반']
b = [80,90,85,95]

# x축 : 학급
# y축 : 점수
plt.bar(a,b)
```

#### 산점도
- scatter : 화면에 산점도를 그리는 함수

```python
# 산점도에 담을 데이터 생성 (x,y의 갯수는 같아야함)
x = [30,25,10,50,25]
y = [9,3,10,6,2]

# x축 : 첫번째 배열에 따라 자동으로 범위가 설정됨
# y축 : 두번째 배열에 따라 자동으로 범위가 설정됨

# 산점도 그리기
plt.scatter(x,y)
```

#### 히스토그램
- hist : 화면에 히스토그램을 그리는 함수 => y축 미 지정시 자동으로 설정

```python
# 히스토그램에 사용할 데이터 만들기
# 시험 점수 100개 만들기 (랜덤함수)
# => randint : 랜덤한 정수를 만드는 함수(시작, 끝, 개수)
scores=np.random.randint(50,101,100)
scores

# 히스토그램 그리기 
# plt.hist(데이터,구간)
plt.hist(scores,bins=10)

# 그래프 꾸미기 => 다른 그래프에도 사용 가능
plt.title('시험 점수 분포') # 그래프 제목
plt.xlabel('시험 점수')    # x축 라벨
plt.ylabel('학생의 수')    # y축 라벨
```

### 8.2 여러 그래프 그리기
- figure(피겨): 흰 도화지  
=>도화지의 사이즈를 설정하거나, 여러 그래프를 한번에 그릴 때 사용    
- axes(엑시스) : 도화지 안에 들어가는 그래프  

```python
# 1부터 10까지 숫자 생성
a = np.arange(1,11)
plt.plot(a)
# plot함수를 쓰면 자동으로 figure(도화지)가 생성된다
```

```python
# 흰 도화지 만들기
# 가로 3인치, 세로 3인치 만큼 크기 설정하기
plt.figure(figsize=(3,3))
# 도화지 위에 그래프 그리기
plt.plot(np.arange(1,11))
```

```python
# 도화지를 만들고 그 위에 그래프 여러개 그리기
# 도화지 만들기
fig = plt.figure(figsize=(10,3))

# 도화지 위에 그래프를 어떤 모양으로 그릴지 생각하기
# 두개의 그래프를 한줄에 표시

# add_subplot(행, 열, 번호)
ax1 = fig.add_subplot(1,2,1) # 1행 2열 중 1번째 
ax2 = fig.add_subplot(1,2,2) # 1행 2열 중 2번째 

# 학생 시험 점수 데이터
students = [1,2,3,4,5] # 학번
scores = [80,90,70,85,95] # 점수

# 첫번째로 선 그래프 그리기
# 위에서 만든 엑시스 사용
ax1.plot(scores)

# 두번째는 막대 그래프 그리기
ax2.bar(students,scores)
```

#### 여러 그래프 그리기2

```python
# subplots : 피겨와 엑시스를 한번에 생성하는 함수
# 위에서 작성했던 방식보다 편함
# subplots(행 크기, 열 크기, 캔버스사이즈)
fig,axs = plt.subplots(1,2,figsize=(10,3))

# 현재 axs는 두개 크기의 배열
# axs[0] : 1행2열의 첫번째 그래프
# axs[1] : 1행2열의 두번째 그래프 
students = [1,2,3,4,5]
scores = [80,90,70,85,95]

# 첫번째 그래프 그리기
axs[0].plot(scores)
# 두번째 그래프 그리기
axs[1].bar(students,scores)
```
