## 공공데이터포털에서 “국토교통부_전국 버스정류장 위치정보”를 검색하세요.
1. 인천 버스정류장에서 정류장명에 ‘마을회관’이 포함된 정류장  
2. 서울 버스정류장에서 정류장명에 ‘상가’가 포함된 정류장  

```python
# 모튤 불러오기
import pandas as pd
import folium 
```

```python
# 인천 버스정류장에서 정류장명에 ‘마을회관’이 포함된 정류장
data = pd.read_csv('국토교통부_전국 버스정류장 위치정보_20241028.csv',encoding='euc-kr')
data

df = data[['정류장명','위도','경도','도시명']]
df

inc=df[df['도시명'] == '인천광역시']


inc1 = inc[inc['정류장명'].str.contains('마을회관')]

print(inc1)

```

```python
# 2.서울 버스정류장에서 정류장명에 ‘상가’가 포함된 정류장
data = pd.read_csv('국토교통부_전국 버스정류장 위치정보_20241028.csv',encoding='euc-kr')
data

df = data[['정류장명','위도','경도','도시명']]

seo=df[df['도시명'] == '서울특별시']


seo1 = seo[seo['정류장명'].str.contains('상가')]

print(seo1)
```

```python
# 그리고 지도 위에 정류장을 표시해주세요.
# 단, 인천 정류장은 빨강, 서울정류장은 파랑으로 표시하세요.
```

```python
# 지도 만들기
map=folium.Map(location =[37.497547, 126.861803], zoom_start=10)
map
```

```python
# 인천정류장 마커표시
inc1.index
inc1.columns

for i in inc1.index :
    name = inc1.loc[i,'정류장명']
    lat = inc1.loc[i,'위도']
    lon = inc1.loc[i,'경도']
    typ = inc1.loc[i,'도시명']
    # print(i, name, typ, lat, lon)

    folium.Marker([lat, lon], tooltip = name, icon = folium.Icon(color = 'red') ).add_to(map)
map
```

```python
# 서울 정류장
seo1.index
seo1.columns

for i in seo1.index :
    name = seo1.loc[i,'정류장명']
    lat = seo1.loc[i,'위도']
    lon = seo1.loc[i,'경도']
    typ = seo1.loc[i,'도시명']
    # print(i, name, typ, lat, lon)

    folium.Marker([lat, lon], tooltip = name, icon = folium.Icon(color = 'blue') ).add_to(map)
map
```
