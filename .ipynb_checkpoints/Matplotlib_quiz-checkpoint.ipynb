```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import rc
rc('font', family='Malgun Gothic')
```

## [문제1]
1. 선 그래프를 그리세요.  
값 : 10부터 20까지  
2. 막대 그래프를 그리세요.  
과일 : ['사과','바나나','포도','오렌지']  
판매량 : [50, 80, 40, 70]  

```python
# 1.선 그래프를 그리세요.
a = np.arange(10,21)

plt.plot(a)
```

```python
# 2.막대 그래프를 그리세요.
a = ['사과','바나나','포도','오렌지']
b = [50, 80, 40, 70]
plt.bar(a,b)
```

3. 산점도 그래프를 그리세요.  
x : [1, 2, 3, 4, 5]  
y : [40, 55, 65, 80, 85]  

4. 히스토그램을 그리세요.  
- 값: 하루 유튜브 시청 시간(0~10시간 사이, 100명)  

```python
# 산점도 그래프를 그리세요.
x = [1, 2, 3, 4, 5]
y = [40, 55, 65, 80, 85]
plt.scatter(x,y)

plt.title('산점도 그래프')
plt.xlabel('공부 시간')
plt.ylabel('시험 점수')
```

```python
# 히스토그램을 그리세요
# 값: 하루 유튜브 시청 시간(0~10시간 사이, 100명)
yt=np.random.randint(0,11,100)
yt

plt.hist(yt,bins=10)

plt.title('유튜브 시청 시간 분포')
plt.xlabel('시청 시간(시간)')
plt.ylabel('사람 수')
```

## [문제2]
1. 하나의 도화지에 2개의 그래프(선 그래프, 막대 그래프) 그리세요.  
데이터 : 과일 판매량  
과일 : ['사과', '바나나', '포도', '오렌지', '복숭아']  
판매 개수 : [50, 80, 40, 70, 60]  

```python
# 하나의 도화지에 2개의 그래프(선 그래프, 막대 그래프) 그리세요.
fig,axs = plt.subplots(1,2,figsize=(10,3))

fr = ['사과', '바나나', '포도', '오렌지', '복숭아']
tt= [80,90,70,85,95]

axs[0].plot(fr,tt)
axs[0].set_title('과일 판매량')
axs[1].bar(fr,tt)
axs[1].set_title('과일 판매량')
```

```python
# 2. 하나의 도화지에 2개의 그래프(선 그래프, 막대 그래프) 그리세요.
#데이터 : 요일별 유튜브 시청 시간
#요일 : ['월', '화', '수', '목', '금']
#시청 시간 : [2, 3, 4, 2, 5]

fig,axs = plt.subplots(1,2,figsize=(10,3))
day = ['월', '화', '수', '목', '금']
time =  [2, 3, 4, 2, 5]

axs[0].plot(day,time)
axs[0].set_title('요일별 유튜브 시청 시간')
axs[1].bar(day,time)
axs[1].set_title('요일별 유튜브 시청 시간')
```

```python
# 3. 하나의 도화지에 2개의 그래프(막대 그래프, 산점도) 그리세요.
# 데이터 : 요일별 독서 시간
# 요일 : ['월', '화', '수', '목', '금’]
# 독서 시간 : [1, 0.5, 2, 1.5, 2.5]
fig,axs = plt.subplots(1,2,figsize=(10,3))
day = ['월', '화', '수', '목', '금']
reading = [1, 0.5, 2, 1.5, 2.5]


axs[0].bar(day,reading)
axs[0].set_title('요일별 독서 시간(막대그래프)')
axs[1].scatter(day,reading)
axs[1].set_title('요일별 독서 시간(산점도)')

```

```python
# 4. 하나의 도화지에 2개의 그래프(선 그래프, 산점도)를 그리세요.
# 데이터 : 도시별 평균 기온(°C)
# 도시 : ['서울', '부산', '대구', '광주', '제주’]
# 평균 기온 : [15.5, 17.8, 16.2, 15.0, 18.3]

fig,axs = plt.subplots(1,2,figsize=(10,3))
city = ['서울', '부산', '대구', '광주', '제주']
c = [15.5, 17.8, 16.2, 15.0, 18.3]

axs[0].plot(city,c)
axs[0].set_title('도시별 평균 기온(선그래프)')
axs[1].scatter(city,c)
axs[1].set_title('도시별 평균 기온(산점도)')
```
