### [문제4]
주민등록인구및세대현황.csv 파일을 불러오세요.  
2025년08월_남자 인구수 컬럼에서 값이 없는 행을 확인하세요.  
2025년08월_남자 인구수 컬럼에서 값이 없는 행의 개수를 구하세요.  
2025년08월_세대당 인구 값이 없는 행을 모두 삭제하세요.  
2025년08월_남자 인구수 값이 비어 있으면 999명으로 대체하세요.  


```python
# 주민등록인구및세대현황.csv 파일을 불러오세요
data = pd.read_csv('0917/주민등록인구및세대현황.csv',encoding='euc-kr')

df = data[['행정구역','남자 인구수','여자 인구수','남여 비율']]
df
```

```python
# 남자 인구수 컬럼에서 값이 없는 행을 확인하세요.
df['남자 인구수'].isnull()
```

```python
# 남자 인구수 컬럼에서 값이 없는 행의 개수를 구하세요.
df['남자 인구수'].isnull().sum()
```

```python
# 세대당 인구 값이 없는 행을 모두 삭제하세요.
data.dropna(subset=['세대당 인구'])
```

```python
# 남자 인구수 값이 비어 있으면 999명으로 대체하세요.
df['남자 인구수'].fillna(999)
```

```python
# 세대당 인구가 없으면0, 남자 인구수 값이 비어 있으면 999명으로 대체하세요.
data.fillna({'세대당 인구': 0,'남자 인구수': 999})
```

### [문제12]  
코로나 확진자 현황데이터  
연도별 총 확진자수를 구하세요  
동별 평균 확진자수를 구하세요  
2021년도 월별 확진자수의 합계를 구하세요  
확진자수가 없는 행의 개수를 구하세요  
동별 확진자수를 구하고, 큰값 순으로 정렬하세요(5개만 출력)  
2022년 확진자수가 가장 많은 동과 확진자수를 구하세요

```python
# 코로나 확진자 현황데이터
df = pd.read_csv('0917/서울특별시 마포구_법정동별_월별_코로나19 확진자 현황.csv',encoding='euc-kr')
df
```

```python
# 연도별 총 확진자수를 구하세요
df.groupby('기준연도').sum()['확진자수']
```

```python
# 동별 평균 확진자수를 구하세요

df = df.dropna()

df = df.astype({'확진자수':'int64'})

df.groupby('법정동')['확진자수'].mean()

# groupby위에 형변환을 해주는 함수를 사용하더라도,
# mean()의 결괏값이 float로 자동 변환 되기 때문에 int로 반환되지x
# int로 바꾸고 싶다면 소수점 이하를 버리거나 반올림하여 변환하는 함수를 사용해 준다
# ex)df.groupby('법정동')['확진자수'].mean().astype('int') or df.groupby('법정동')['확진자수'].mean().round().astype('int')
```

```python
# 2021년도 월별 확진자수의 합계를 구하세요
df[df['기준연도'] == 2021].groupby('기준월').sum()['확진자수']
```

```python
# 확진자수가 없는 행의 개수를 구하세요
df['확진자수'].isnull().sum()
```

```python
# 동별 확진자수를 구하고, 큰값 순으로 정렬하세요(5개만 출력)
df.groupby('법정동')['확진자수'].sum().sort_values(ascending = False).head()
```

```python
# 2022년 확진자수가 가장 많은 동과 확진자수를 구하세요
df[df['기준연도'] == 2022].groupby('법정동').sum()['확진자수']['성산동'] # =>25571
```
