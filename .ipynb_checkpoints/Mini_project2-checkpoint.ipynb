## 공공데이터포털에서 “국토교통부_전국 버스정류장 위치정보”를 검색하세요.
1. 인천 버스정류장에서 정류장명에 ‘마을회관’이 포함된 정류장  
2. 서울 버스정류장에서 정류장명에 ‘상가’가 포함된 정류장  

```python
# 모튤 불러오기
import pandas as pd
import folium 
```

```python
# 인천 버스정류장에서 정류장명에 ‘마을회관’이 포함된 정류장
data = pd.read_csv('국토교통부_전국 버스정류장 위치정보_20241028.csv',encoding='euc-kr')
data

df = data[['정류장명','위도','경도','도시명']]
df

inc=df[df['도시명'] == '인천광역시']


df1 = inc[inc['정류장명'].str.contains('마을회관')]

print(df1)

```

```python
# 2.서울 버스정류장에서 정류장명에 ‘상가’가 포함된 정류장
data = pd.read_csv('국토교통부_전국 버스정류장 위치정보_20241028.csv',encoding='euc-kr')
data

df = data[['정류장명','위도','경도','도시명']]

seo=df[df['도시명'] == '서울특별시']


df2 = seo[seo['정류장명'].str.contains('상가')]

print(df2)
```

```python
# 그리고 지도 위에 정류장을 표시해주세요.
# 단, 인천 정류장은 빨강, 서울정류장은 파랑으로 표시하세요.
```

필요한 데이터가 여기저기 흩어져 있음 => 데이터 합치기
- merge : 조건이 맞으면 연결  
- concat: 그냥 이어붙이기  
    -axes=0 : 위,아래로 연결  
    -axes=1 : 옆으로 연결  

```python
bus = pd.concat([df1, df2])
bus
```

```python
# 지도 만들기
map=folium.Map(location =[37.497547, 126.861803], zoom_start=10)
map
```

```python
# 정류장 마커표시
# 정류장 마커 추가
for i in bus.index:
    name = bus.loc[i, '정류장명']
    lat = bus.loc[i, '위도']
    lon = bus.loc[i, '경도']
    city = bus.loc[i, '도시명']

    # 지역에 따라 마커 색 변경
    if city == '인천광역시':
        color = "red"
    elif city == '서울특별시':
        color = "blue"
    
    folium.Marker( [lat, lon], popup=name, tooltip=name, icon=folium.Icon(color=color) ).add_to(map)
map
```
