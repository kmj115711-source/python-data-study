## 6.2 없는 데이터 nan처리 하기
- nan : csv에서 빈셀을 의미 값이 없다

빈값이 있는지 확인하기
- isnull : 빈값이 있는지 확인  
- isnull_sum : 빈값의 개수  

```python
data = pd.read_csv('0917/서울특별시 마포구_법정동별_월별_코로나19 확진자 현황.csv', encoding='euc-kr')
df = data[['기준연도','기준월','법정동','확진자수']]
```

```python
df['확진자수']
# 확진자수에 빈값이 있는지 확인
# 각 행에 대해서 검사 => bool시리즈 반환
print(df['확진자수'].isnull())
# 빈값 개수 세기 = bool =>int
# false : 0, true : 1
df['확진자수'].isnull().sum()
```

### 빈값 삭제하기
- dropna : 빈값이 있는 행을 삭제
- dropna(subset=['컬럼']) : 빈값이 있는지 특정 컬럼만 검사하여, 삭제

```python
# 빈 값이 있는 행을 모두 삭제
df.dropna() 
# 확진자수만 검사하여, 행 삭제
df.dropna(subset=['확진자수'])
# => dropna는 결과만 보여주고, 원본은 그대로 
```

### 빈값을 다른 값으로 대체  
- fillna(새로운값)
- fillna({컬럼1:새로운값, 컬럼2:새로운값}) : 여러 컬럼을 한번에 처리할때

```python
df
# 법정동에 빈값이 있으면 '모름'으로 표시
df['법정동'].fillna('모름')

# 확진자수에 빈값이 있으면 0으로 표시
df['확진자수'].fillna(0)

# => 원본은 그대로 

# 여러 컬럼을 한번에 교체
# 법정동과 확진자수의 빈값을 한번에 처리 => 원본 그대로 
df.fillna({'법정동': '모름','확진자수':0})

```

```python
# 변경된 내용을 원본에 넣기
df['법정동'] = df['법정동'].fillna('모름')
df['확진자수'] = df['확진자수'].fillna(0)
df
```

```python
df = df.fillna({'법정동': '모름','확진자수':0})
df
```

## 6.3 조건검색
특정 조건에 맞는 데이터 찾기

```python
# 코로나 확진자 현황 데이터 가져오기
data = pd.read_csv('0917/서울특별시 마포구_법정동별_월별_코로나19 확진자 현황.csv', encoding='euc-kr')
df = data[['기준연도','기준월','법정동','확진자수']]
df
```

```python
# 확진자수가 3000명 이상인 행 찾기
df[df['확진자수'] >= 3000]
```

```python
# 확진자수가 제일 많은 행 찾기
# 내림차순으로 정렬
df.sort_values('확진자수', ascending = False).head()
```

### 조건 검색을 위한 시리즈의 문자열 함수
- df[' '].str : 문자열 함수

```python
# 1.법정동이 '공덕'으로 시작하는 행 찾기 => startswith는 시작!
# => 해당 컬럼에 빈값이 있으면 잘못된 결과가 나올 수 있음
# => 2.삭제, 교체 후 함수계산
df['법정동'] = df['법정동'].fillna('모름')
df
# df['법정동'].dtype # => object
# 3.인덱스로 묶어 주기
df[df['법정동'].str.startswith('공덕')].head()
```

```python
# 법정동에 '고산'을 포함하고 있는 행 찾기 => contains는 포함!
df[df['법정동'].str.contains('고산')].head()
```

```python
# 법정동에 '흥동'으로 끝나는 행 찾기 => endswith 끝!
df[df['법정동'].str.endswith('흥동')].head()
```

## 6.4 집계 및 변형
- groupby : 특정 컬럼을 기준으로 행 묶기
  데이터를 그룸화 한 후에 합계, 평균을 구할 수 있다

```python
# 코로나 데이터 가져오기
data = pd.read_csv('0917/서울특별시 마포구_법정동별_월별_코로나19 확진자 현황.csv', encoding='euc-kr')
df = data[['기준연도','기준월','법정동','확진자수']]
df
```

```python
# 동별 총 확진자수 구하기
# 그룹핑을 하면 기준컬럼이 인덱스로 바뀜
result = df.groupby('법정동').sum()
result['확진자수']

# 여기서 공덕동의 값만 확인
# type(result['확진자수']) # => Series

result['확진자수']['공덕동'] # 20999.0
```

- type : 컬럼의 객체 타입 확인
- .dtype : 원소의 타입 확인
- info : 데이터프레임의 모든 컬럼 정보를 확인

```python
# 데이터프레임에서 열 하나 꺼내기
df['확진자수']
# 열의 자료 구조
print(type(df['확진자수']))
# 원소의 자료형
print(df['확진자수'].dtype)
```

```python
# 데이터프레임의 모든 컬럼 정보
df.info()
```

### 컬럼의 타입 변경하기
- astype({컬럼 : 새타입})

```python
# 코로나 확진자수는 사람수인데 float으로 표시됨
# float => int 형변환
# =>nan값 때문에 오류 => 삭제 또는 교체 후 사용

df = df.dropna()

df = df.astype({'확진자수':'int64'}) # => astype도 원본엔 영향 x =>원본파일에 꼭 덮어쓰기해줘야함
df
```
