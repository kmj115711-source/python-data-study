```python
# 하나의 도화지에 2개의 그래프(선 그래프, 막대 그래프) 그리세요.
fig,axs = plt.subplots(1,2,figsize=(10,3))

fr = ['사과', '바나나', '포도', '오렌지', '복숭아']
tt= [80,90,70,85,95]

axs[0].plot(fr,tt)
axs[0].set_title('과일 판매량')
axs[1].bar(fr,tt)
axs[1].set_title('과일 판매량')
```

```python
# 2. 하나의 도화지에 2개의 그래프(선 그래프, 막대 그래프) 그리세요.
#데이터 : 요일별 유튜브 시청 시간
#요일 : ['월', '화', '수', '목', '금']
#시청 시간 : [2, 3, 4, 2, 5]

fig,axs = plt.subplots(1,2,figsize=(10,3))
day = ['월', '화', '수', '목', '금']
time =  [2, 3, 4, 2, 5]

axs[0].plot(day,time)
axs[0].set_title('요일별 유튜브 시청 시간')
axs[1].bar(day,time)
axs[1].set_title('요일별 유튜브 시청 시간')
```

```python
# 3. 하나의 도화지에 2개의 그래프(막대 그래프, 산점도) 그리세요.
# 데이터 : 요일별 독서 시간
# 요일 : ['월', '화', '수', '목', '금’]
# 독서 시간 : [1, 0.5, 2, 1.5, 2.5]
fig,axs = plt.subplots(1,2,figsize=(10,3))
day = ['월', '화', '수', '목', '금']
reading = [1, 0.5, 2, 1.5, 2.5]


axs[0].bar(day,reading)
axs[0].set_title('요일별 독서 시간(막대그래프)')
axs[1].scatter(day,reading)
axs[1].set_title('요일별 독서 시간(산점도)')

```

```python
# 4. 하나의 도화지에 2개의 그래프(선 그래프, 산점도)를 그리세요.
# 데이터 : 도시별 평균 기온(°C)
# 도시 : ['서울', '부산', '대구', '광주', '제주’]
# 평균 기온 : [15.5, 17.8, 16.2, 15.0, 18.3]

fig,axs = plt.subplots(1,2,figsize=(10,3))
city = ['서울', '부산', '대구', '광주', '제주']
c = [15.5, 17.8, 16.2, 15.0, 18.3]

axs[0].plot(city,c)
axs[0].set_title('도시별 평균 기온(선그래프)')
axs[1].scatter(city,c)
axs[1].set_title('도시별 평균 기온(산점도)')
```

### [문제5]

```python
# 1. 요일별 평균 기온 데이터를 DataFrame으로 만들고, 다음과 같이 그래프를 그리세요.
# 첫 번째 서브플롯: 선 그래프
# 두 번째 서브플롯: 산점도

df = pd.DataFrame({'요일' :['월','화','수','목','금','토','일'],
                  '기온' : [12,14,13,15,17,18,16]})

fig, axs = plt.subplots(1,2, figsize= (10,3))

sns.lineplot(data = df, x = '요일', y = '기온', ax=axs[0])
axs[0].set_title('요일별 기온(선)')

sns.scatterplot(data = df, x = '요일', y = '기온', ax=axs[1])
axs[1].set_title('요일별 기온(산점도)')
```

```python
# 2. 과일별 판매량 데이터를 DataFrame으로 만들고, 다음과 같이 그래프를 그리세요.
# 첫 번째 서브플롯: 막대 그래프
# 두 번째 서브플롯: 선 그래프
# 세 번째 서브플롯: 산점도
df = pd.DataFrame({'과일' : ['사과','바나나','포도','딸기'], 
                  '판매량' : [120, 80, 150, 100]})

fig, axs = plt.subplots(1,3, figsize= (10,3))

sns.barplot(data = df, x = '과일', y = '판매량', ax=axs[0])
axs[0].set_title('과일별 판매량(막대)')

sns.lineplot(data = df, x = '과일', y = '판매량', ax=axs[1])
axs[1].set_title('과일별 판매량(선)')

sns.scatterplot(data = df, x = '과일', y = '판매량', ax=axs[2])
axs[2].set_title('과일별 판매량(산점도)')

plt.tight_layout()
```

```python
# 3. 학생 시험 점수 데이터를 DataFrame으로 만들고, 다음과 같이 그래프를 그리세요.
# 첫 번째 서브플롯: 산점도
# 두 번째 서브플롯: 막대 그래프
# 세 번째 서브플롯: 선 그래프

df = pd.DataFrame({'공부시간' : [2, 4, 6, 3, 5, 7, 8], 
                  '점수' : [55, 65, 75, 60, 70, 85, 90]})

fig, axs = plt.subplots(3,1, figsize= (3,10))

sns.scatterplot(data = df, x = '공부시간', y = '점수', ax=axs[0])
axs[0].set_title('공부시간-점수(산점도)')

sns.barplot(data = df, x = '공부시간', y = '점수', ax=axs[1])
axs[1].set_title('공부시간-점수(막대)')

sns.lineplot(data = df, x = '공부시간', y = '점수', ax=axs[2], marker='o')
axs[2].set_title('공부시간-점수(선)')

plt.tight_layout()

```

```python
# 4. 과일 판매량과 학생 시험 점수 데이터를 DataFrame으로 만들고, 다음과 같이 그래프를 그리세요.
#첫 번째 서브플롯(왼쪽 위)에는 과일 판매량 막대 그래프
#두 번째 서브플롯(오른쪽 위)에는 과일 판매량 선 그래프
#세 번째 서브플롯(왼쪽 아래)에는 학생 시험 점수 막대 그래프
#네 번째 서브플롯(오른쪽 아래)에는 학생 시험 점수 산점도

df1 = pd.DataFrame({'과일' : ['사과','바나나','포도','딸기'], 
                  '판매량' : [120, 80, 150, 100]})

df2 = pd.DataFrame({'이름' : ['영희','철수','민수','지연'], 
                  '점수' : [85, 90, 80, 100]})


fig, axs = plt.subplots(2,2, figsize= (8,8))
axs

sns.barplot(data = df1, x = '과일', y = '판매량', ax=axs[0,0])
axs[0,0].set_title('과일별 판매량(막대)')

sns.lineplot(data = df1, x = '과일', y = '판매량', ax=axs[0,1], marker='o')
axs[0,1].set_title('과일별 판매량(선)')


sns.barplot(data = df2, x = '이름', y = '점수', ax=axs[1,0])
axs[1,0].set_title('학생별 점수(막대)')
plt.xlabel('학생')

sns.scatterplot(data = df2, x = '이름', y = '점수', ax=axs[1,1])
axs[1,1].set_title('학생별 점수(산점도)')
plt.xlabel('학생')


plt.tight_layout()
```
