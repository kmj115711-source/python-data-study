## 5.3 데이터베이스 연결하기  
설치형 데이터베이스 : mysql, oracle..  
파일형 데이터베이스 : sqlite..  
파일형 데이터베이스는 파일만 있으면 어디서든 열 수 있다  

### 데이터 베이스 연결 및 테이블 생성

```python
# 모듈 불러오기
import sqlite3
```

```python
# 데이터베이스 연결 또는 생성
# connect(데이터베이스 이름)
con = sqlite3.connect('mydata.sqlite')
```

```python
# 학생의 정보를 저장할 테이블 생성
# 데이터베이스 명령 : sql(쿼리)
# create 
# insert select uqdate delete
con.execute('create table student(id integer, name text, age integer)')
# 테이블에 데이터 추가
con.execute("insert into student values(1,'둘리',25)")
con.execute("insert into student values(2,'도우너',30)")
con.commit()
```

### 테이블에 있는 데이터 조회

```python
# 조회 쿼리 실행
cur = con.execute('select * from student')
# 쿼리결과 꺼내기
rows = cur.fetchall()
for tup in rows:
    print(tup)
```

### 단건 데이터 조회

```python
# 1번 학생만 조회
cur = con.execute('select * from student where id =1')
# 쿼리 결과 확인
row = cur.fetchone()
row
```
